# Build model - CNN multi
____________________

```
def CNNmulti(vocab_size, max_len, embedding, output_size, kernel_sizes = [4, 6, 8]):
  #### INITIALIZER
  initializer = initializers.GlorotUniform(seed=settings['RANDOM_STATE'])

  #### INPUT
  inputs = Input(shape=(max_len,))

  #### EMBEDDING LAYER
  embedding = Embedding(input_dim=vocab_size, output_dim=embedding.shape[1], weights=[embedding], input_length=max_len, trainable=False)(inputs)

  spatial = SpatialDropout1D(0.2)(embedding)

  #### CONVOLUTIONAL LAYERS
  # define first
  conv1 = Conv1D(32, kernel_sizes[0], activation='relu', kernel_initializer=initializer)(spatial)
  drop1 = Dropout(0.3)(conv1)
  pool1 = MaxPooling1D(2)(drop1)
  flat1 = Flatten()(pool1)

  # define second
  conv2 = Conv1D(32, kernel_sizes[1], activation='relu', kernel_initializer=initializer)(embedding)
  drop2 = Dropout(0.3)(conv2)
  pool2 = MaxPooling1D(2)(drop2)
  flat2 = Flatten()(pool2)

  # define third
  conv3 = Conv1D(32, kernel_sizes[2], activation='relu', kernel_initializer=initializer)(embedding)
  drop3 = Dropout(0.3)(conv3)
  pool3 = MaxPooling1D(2)(drop3)
  flat3 = Flatten()(pool3)

  # merge all
  merged = Concatenate()([flat1, flat2, flat3])

  #### FULLY CONNECTED LAYERS
  dense1 = Dense(128, activation='relu', kernel_initializer=initializer)(merged)
  dropout = Dropout(0.3)(dense1)

  outputs = Dense(output_size, activation='sigmoid')(dropout)

  # compile model
  model = Model(inputs=inputs, outputs=outputs)
  model.compile(loss='binary_crossentropy', optimizer='adam', metrics=[metrics.Precision(), metrics.AUC(multi_label=True)])
  print(model.summary())
  return model
```
